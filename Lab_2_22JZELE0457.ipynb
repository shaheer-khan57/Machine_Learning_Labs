## 2. Pandas Essentials

```python
import pandas as pd
print("Pandas version:", pd.__version__)
```

```python
# Sample Data
data = {
    'Name': ['Shaheer Khan', 'Ali Ahmed', 'Sara Khan', 'Usman Ali', 'Fatima Noor'],
    'Marks': [85, 72, 91, 65, 78],
    'Age': [20, 21, 19, 22, 20]
}
df = pd.DataFrame(data)
df
```

### **Pandas Mini Exercises**

```python
# 1. Add Grade column
def get_grade(marks):
    if marks >= 85: return 'A'
    elif marks >= 70: return 'B'
    elif marks >= 60: return 'C'
    else: return 'F'

df['Grade'] = df['Marks'].apply(get_grade)
df
```

```python
# 2. Sort by Marks
df_sorted = df.sort_values(by='Marks', ascending=False)
df_sorted
```

## 3. Matplotlib Essentials

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### **Matplotlib Mini Exercises**

```python
# 1. Line plot - Temperature for 7 days
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temp = [32, 34, 31, 33, 35, 36, 33]

plt.figure(figsize=(8,5))
plt.plot(days, temp, marker='o', color='blue', linewidth=2, label='Temperature')
plt.title('Daily Temperature for 7 Days')
plt.xlabel('Days')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 2. Bar chart - Favorite fruits
fruits = ['Apple', 'Banana', 'Mango', 'Orange', 'Grapes']
students = [25, 30, 18, 22, 15]

plt.figure(figsize=(8,5))
plt.bar(fruits, students, color=['red','yellow','orange','blue','purple'])
plt.title('Favorite Fruits Among Students')
plt.xlabel('Fruits')
plt.ylabel('Number of Students')
plt.grid(axis='y')
plt.show()
```

```python
# 3. Scatter plot - Height vs Weight
height = [160, 165, 170, 155, 175, 168, 172, 158, 180, 162]
weight = [55, 60, 70, 50, 75, 65, 68, 52, 80, 58]

plt.figure(figsize=(8,6))
plt.scatter(height, weight, color='green', s=100, alpha=0.7)
plt.title('Student Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

```python
# 4. Pie chart - Daily activities
activities = ['Study', 'Sleep', 'Play', 'Exercise', 'Others']
hours = [6, 8, 4, 2, 4]

plt.figure(figsize=(8,6))
plt.pie(hours, labels=activities, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title('Time Spent on Daily Activities')
plt.show()
```

```python
# 5. Histogram - Marks of 20 students
marks = np.random.randint(40, 95, 20)

plt.figure(figsize=(8,5))
plt.hist(marks, bins=8, color='purple', edgecolor='black', alpha=0.7)
plt.title('Distribution of Student Marks')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.grid(True, axis='y')
plt.show()
```

```python
# 6. Subplots: Line + Bar
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Line plot
axes[0].plot(days, temp, marker='o', color='blue')
axes[0].set_title('Temperature Trend')
axes[0].set_xlabel('Days')
axes[0].set_ylabel('Temp (°C)')
axes[0].grid(True)

# Bar plot
axes[1].bar(fruits, students, color='orange')
axes[1].set_title('Favorite Fruits')
axes[1].set_xlabel('Fruits')
axes[1].set_ylabel('Students')
axes[1].grid(axis='y')

plt.tight_layout()
plt.show()
```
