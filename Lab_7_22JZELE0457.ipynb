# Lab 7: Corn Leaf Disease Dataset Preparation

## Objective
Split the original PlantVillage corn leaf dataset into structured **Training**, **Validation**, and **Test** sets for binary and multi-class classification tasks.

## Original Dataset Structure
- `Corn_(maize)___healthy`
- `Corn_(maize)___Cercospora_leaf_spot Gray_leaf_spot`
- `Corn_(maize)___Common_rust_`
- `Corn_(maize)___Northern_Leaf_Blight`

## Output Structure (Created by this notebook)
```
comm/
├── train/
│   ├── healthy/           (400 images)
│   ├── Cercospora/        (400 images)
│   ├── common_rust/       (400 images)
│   └── northern_leaf_blight/ (400 images)
├── validation/
│   ├── healthy/           (63 images)
│   ├── Cercospora/        (63 images)
│   ├── common_rust/       (63 images)
│   └── northern_leaf_blight/ (63 images)
└── test/
    ├── healthy/           (50 images)
    ├── Cercospora/        (50 images)
    ├── common_rust/       (50 images)
    └── northern_leaf_blight/ (50 images)
```

> **Note:** The splits (400/63/50) are fixed per the lab instructions. Images are renamed sequentially and selected via OS-level ordering from the source folders.

````python
import os
import shutil

# Define source and destination root paths
original_dataset_dir = r'C:\Users\PMYLS\Documents\MachineLearningLaB\Corn'
base_dir = r'C:\Users\PMYLS\Documents\MachineLearningLaB\comm'
````

## 1. Create Base Directories
Creates `train/`, `validation/`, and `test/` at the destination. Uses `exist_ok=True` to safely rerun.

````python
os.makedirs(base_dir, exist_ok=True)

train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')

os.makedirs(train_dir, exist_ok=True)
os.makedirs(validation_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

print("Base directories created successfully.")
````

## 2. Create Class Subdirectories
Creates separate folders for each class (`healthy`, `Cercospora`, `common_rust`, `northern_leaf_blight`) inside train/val/test.

````python
# Define class names (matching folder naming conventions)
classes = ['healthy', 'Cercospora', 'common_rust', 'northern_leaf_blight']

# Generate paths for each split and class
for split in ['train', 'validation', 'test']:
    split_path = os.path.join(base_dir, split)
    for cls in classes:
        os.makedirs(os.path.join(split_path, cls), exist_ok=True)

print("All class subdirectories are ready.")
````

## 3. Verify Source Folders
Lists all folders present in the original dataset directory to ensure paths match.

````python
print("Found source folders:")
for item in os.listdir(original_dataset_dir):
    print(f"  {item}")
````

## 4. Copy Training Images (400 per class)
Takes the **first 400** files from each source folder, renames them sequentially, and copies them to the corresponding train subfolder.

> **What is `i`?** 
> - In `range(400)`, `i` is the loop counter used both as an index to access `src_list[i]` and as a number to generate the new filename (e.g., `healthy.0.jpg`).

````python
# Mapping source folder names to destination class names
source_map = {
    'healthy': 'Corn_(maize)___healthy',
    'Cercospora': 'Corn_(maize)___Cercospora_leaf_spot Gray_leaf_spot',
    'common_rust': 'Corn_(maize)___Common_rust_',
    'northern_leaf_blight': 'Corn_(maize)___Northern_Leaf_Blight'
}

TRAIN_COUNT = 400

for cls, src_folder_name in source_map.items():
    src_path = os.path.join(original_dataset_dir, src_folder_name)
    src_list = os.listdir(src_path)  # OS order (varies, but consistent per run)
    
    dst_dir = os.path.join(train_dir, cls)
    
    # Generate new filenames: class_name.0.jpg, class_name.1.jpg, ...
    fnames = [f'{cls}.{i}.jpg' for i in range(TRAIN_COUNT)]
    
    for i in range(TRAIN_COUNT):
        src = os.path.join(src_path, src_list[i])
        dst = os.path.join(dst_dir, fnames[i])
        shutil.copyfile(src, dst)
    
    print(f"Copied {TRAIN_COUNT} {cls} images to train.")
````

## 5. Copy Validation Images (63 per class)
Takes the **next 63** files (indices 400 to 462) from each source folder, renames them sequentially, and copies to validation.

````python
VAL_COUNT = 63
VAL_START = 400  # start after the training set
VAL_END = VAL_START + VAL_COUNT  # 463

for cls, src_folder_name in source_map.items():
    src_path = os.path.join(original_dataset_dir, src_folder_name)
    src_list = os.listdir(src_path)
    
    dst_dir = os.path.join(validation_dir, cls)
    
    # Validation filenames: class_name.0.jpg to class_name.62.jpg
    fnames = [f'{cls}.{i}.jpg' for i in range(VAL_COUNT)]
    
    for i in range(VAL_START, VAL_END):
        src = os.path.join(src_path, src_list[i])
        dst = os.path.join(dst_dir, fnames[i - VAL_START])
        shutil.copyfile(src, dst)
    
    print(f"Copied {VAL_COUNT} {cls} images to validation.")
````

## 6. Copy Test Images (50 per class)
Takes the **next 50** files (indices 463 to 512) from each source folder, renames them sequentially, and copies to test.

````python
TEST_COUNT = 50
TEST_START = 463  # start after the validation set
TEST_END = TEST_START + TEST_COUNT  # 513

for cls, src_folder_name in source_map.items():
    src_path = os.path.join(original_dataset_dir, src_folder_name)
    src_list = os.listdir(src_path)
    
    dst_dir = os.path.join(test_dir, cls)
    
    # Test filenames: class_name.0.jpg to class_name.49.jpg
    fnames = [f'{cls}.{i}.jpg' for i in range(TEST_COUNT)]
    
    for i in range(TEST_START, TEST_END):
        src = os.path.join(src_path, src_list[i])
        dst = os.path.join(dst_dir, fnames[i - TEST_START])
        shutil.copyfile(src, dst)
    
    print(f"Copied {TEST_COUNT} {cls} images to test.")
````

## 7. Verify Final Dataset Sizes
Counts the number of images in each split/class to confirm everything copied correctly.

````python
print("\n--- Dataset Verification ---\n")

for split_name, split_path in [('TRAIN', train_dir), ('VALIDATION', validation_dir), ('TEST', test_dir)]:
    print(f"{split_name}:")
    for cls in classes:
        cls_path = os.path.join(split_path, cls)
        count = len(os.listdir(cls_path)) if os.path.exists(cls_path) else 0
        print(f"  {cls}: {count} images")
    print()

print("All datasets are ready for model training!")
````
